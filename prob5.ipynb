# Pie Chart

```python
a.columns.values
```

```python
labels = ['Rest BP', 'Cholestrol', 'heart rate']
sizes = [a['trestbps'].mean(), a['chol'].mean(), a['thalach'].mean()]
```

```python
colors = ['Red', 'Blue', 'Green']
```

```python
plt.pie(sizes, colors = colors, labels = labels, autopct = '%1.1f%%', startangle = 100)
plt.title('Average Heart health')
```

# Bar Plot

```python
import numpy as np
```

```python
x = a.iloc[1:101,0]
y = a.iloc[1:101,7]
plt.bar(x, y, align = 'center', alpha = 0.5)
plt.xticks()
plt.ylabel('max Heart Rate')
plt.xlabel('Age')
plt.title('Age vs Max. Heart Rate')
plt.show()
```

# Histogram

```python
hist = sns.histplot(data = a, x = 'age')
hist.set(xlabel = 'Age', ylabel = 'Values', title = 'Histogram for Age')
```

# ScatterPlot

```python
s=a.iloc[1:91,0]
t=a.iloc[1:91,7]
plt.scatter(s,t)
plt.title("Age vs Max Heart Rate")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.show()
```
