# Pie Chart

```python
labels=['Temp','Rel. Humidity','Wind','Rain']
sizes=[a['temp'].mean(),a['RH'].mean(),a['wind'].mean(),a['rain'].mean()]
```

```python
colors = ['Red','Green','Blue','Yellow']
explode = [0,0,0,0]
```

```python
plt.pie(sizes, explode=explode, labels = labels, colors = colors, startangle = 140)
plt.title('Avg weather')
```

# Histogram

```python
import numpy as np
```

```python
hist = sns.histplot(data = a, x='temp')
hist.set(xlabel = 'Temperature', ylabel = 'Value', title = 'Histogram for temperature')
```

# ScatterPlot

```python
s=a.iloc[1:91,10]
t=a.iloc[1:91,8]
plt.scatter(s,t)
plt.title('Wind vs Temperature')
plt.xlabel('Wind')
plt.ylabel('Temperature')
plt.show()
```

# Box Plot

```python
a.boxplot(by='month',column=['temp'], grid=False)
plt.show()
```
